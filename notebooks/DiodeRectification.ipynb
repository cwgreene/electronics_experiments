# Diode Rectification

We investigate the impact of a pure rectification on the frequency spectrum of a signal.

```python
import numpy as np
from matplotlib import pyplot as plt 

t = np.arange(0,2*np.pi,.01)
def rectify(f):
    return f * (f > 0)
```

```python
plt.plot(rectify(np.sin(t)))
```

```python
data_signal = 1 + np.sin(10*t) + 1 + np.sin(20*t)
carrier_signal = np.sin(200*t)
full_signal = data_signal*carrier_signal
plt.plot(full_signal)
plt.plot(data_signal) # desired signal
```

```python
plt.plot(rectify((full_signal)))
plt.plot(data_signal)
```

```python
plt.plot(np.fft.fft(data_signal)[:50]) # big ol DC spike
```

```python
plt.plot(range(1,50), np.fft.fft(data_signal)[1:50]) # focus on frequency at 10
```

```python
plt.plot(np.fft.fft(full_signal)[:400])
# higlihgt side bands
plt.axvline(188,color='red') # slightly off from 190 and 210 to make it clear
plt.axvline(212,color='red')
```

```python
plt.plot(np.fft.fft(rectify(full_signal))[:400])
# higlihgt side bands
plt.axvline(188,color='red') # slightly off from 190 and 210 to make it clear
plt.axvline(212,color='red')
plt.axvline(8,color='green') # and there's our signal!
```
